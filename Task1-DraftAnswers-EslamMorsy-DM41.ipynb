```python
def removeDigits(text):
    trial = ''.join(str(v) for v in text if not v.isdigit())
    return trial
```

```python
sCharacters = ['@','#','_','!','$','%','^','&','*','(',')','<','>','?','/','~']
def beforeSpecialCharacter(text):
    for ltr in text:
        if ltr in sCharacters:
            position = text.index(ltr)
            return text[0:position]
    return text
```
